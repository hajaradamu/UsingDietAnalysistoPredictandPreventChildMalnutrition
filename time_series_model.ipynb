```python
timeseries_df.Time_period = timeseries_df.Time_period.astype(str)
```

```python
timeseries_df.info()
```

```python
# Converting FullDate column into DateTime
timeseries_df.Time_period = pd.to_datetime(timeseries_df["Time_period"])
```

```python
timeseries_df.info()
```

```python
timeseries_df
```

```python
timeseries_df.set_index( "Time_period" , inplace= True )
```

```python
from prophet import Prophet
```

```python
def create_fp_df(burden):
  fp_df = timeseries_df[[burden]].copy()
  fp_df.reset_index(inplace = True)
  fp_df.rename(columns = {'Time_period':'ds'}, inplace = True)
  fp_df.rename(columns = {burden:'y'}, inplace = True)
  return fp_df
```

```python
fp_df = create_fp_df('Stunting')
```

```python
# Initialize the Model
model=Prophet()
```

```python
model.fit(fp_df)
```

```python
# Predicting next 5 years
future_dates=model.make_future_dataframe(5, freq='A')
```

```python
future_dates.tail()
```

```python
prediction=model.predict(future_dates)
```

```python
### plot the predicted projection
model.plot(prediction,include_legend=True)
plt.title("Stunting")
```
